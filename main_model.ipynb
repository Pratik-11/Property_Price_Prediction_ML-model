```python
d = {'no':0,'yes':1,'unfurnished':0,'semi-furnished':1,'furnished':2}
df = df.replace(d)
```

```python
import math
median_area = math.floor(df.area.median())
median_area

df.area = df.area.fillna(median_area)

median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms

df.bedrooms = df.bedrooms.fillna(median_bedrooms)

median_bathrooms = math.floor(df.bathrooms.median())
median_bathrooms

df.bathrooms = df.bathrooms.fillna(median_bathrooms)

median_stories = math.floor(df.stories.median())
median_stories

df.stories = df.stories.fillna(median_stories)

median_mainroad = math.floor(df.mainroad.median())
median_mainroad

df.mainroad = df.mainroad.fillna(median_mainroad)

median_guestroom = math.floor(df.guestroom.median())
median_guestroom

df.guestroom = df.guestroom.fillna(median_guestroom)

median_basement = math.floor(df.basement.median())
median_basement

df.basement = df.basement.fillna(median_basement)

median_hotwaterheating = math.floor(df.hotwaterheating.median())
median_hotwaterheating

df.hotwaterheating = df.hotwaterheating.fillna(median_hotwaterheating)

median_airconditioning = math.floor(df.airconditioning.median())
median_airconditioning

df.airconditioning = df.airconditioning.fillna(median_airconditioning)

median_parking = math.floor(df.parking.median())
median_parking

df.parking = df.parking.fillna(median_parking)

median_prefarea = math.floor(df.prefarea.median())
median_prefarea

df.prefarea = df.prefarea.fillna(median_prefarea)

median_furnishingstatus = math.floor(df.furnishingstatus.median())
median_furnishingstatus

df.furnishingstatus = df.furnishingstatus.fillna(median_furnishingstatus)
```
